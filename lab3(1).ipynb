```python
def find_bread_price(url):
    """Функція для пошуку ціни на білий хліб"""
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
    try:
        r = requests.get(url, headers=headers, timeout=10)
        if r.status_code != 200:
            return None
        soup = BSoup(r.text, 'html5lib')
        for row in soup.find_all('tr'):
            if 'Bread (500 g Loaf)' in row.text or 'Fresh White Bread' in row.text:
                cols = row.find_all('td')
                if len(cols) >= 2:
                    price_text = cols[1].text.replace('€', '').strip()
                    if price_text and price_text != '?':
                        try:
                            return float(price_text)
                        except:
                            pass
        return None
    except:
        return None

print("✅ Функцію find_bread_price створено")
```

```python
DELAY_MIN = 2
DELAY_MAX = 4
BATCH_SIZE = 10
LONG_PAUSE = 30

results = []
total = len(cities)

for i, city in enumerate(cities):
    url1 = f"https://www.numbeo.com/food-prices/in/{city}-France"
    url1 = url1.replace(' ', '-').replace(',', '')
    price = find_bread_price(url1)
    
    if price is None:
        url2 = f"https://www.numbeo.com/food-prices/in/{city}"
        url2 = url2.replace(' ', '-').replace(',', '')
        price = find_bread_price(url2)
    
    if price:
        results.append({'city': city, 'price': price})
        print(f"{city:30} {price:6.2f} €")
    
    if i < total - 1:
        time.sleep(random.uniform(DELAY_MIN, DELAY_MAX))
    
    if (i + 1) % BATCH_SIZE == 0 and i < total - 1:
        time.sleep(LONG_PAUSE)
```

```python
if results:
    # Знаходимо мінімальну та максимальну ціну
    min_price = min(results, key=lambda x: x['price'])
    max_price = max(results, key=lambda x: x['price'])
    
    print(f"\n{min_price['city']} {min_price['price']:.2f}€")
    print(f"{max_price['city']} {max_price['price']:.2f}€")
else:
    print("Немає даних")
```
