```python
import requests
from bs4 import BeautifulSoup as BSoup
import re

print("Бібліотеки завантажено")
```

```python
url = 'https://en.cppreference.com/w/cpp/utility/any'
r = requests.get(url)
print(f"Статус: {r.status_code}")

soup = BSoup(r.text, 'html.parser')
print("Сторінку any завантажено")
```

```python
def get_cpp_items(section_title):
    """Функція для пошуку елементів по заголовку секції"""
    header = soup.find(lambda tag: tag.name in ["h2", "h3"] and section_title in tag.text)
    
    if not header:
        return []
    
    table = header.find_next_sibling("table", class_="t-dsc-begin")
    
    if not table:
        return []
    
    items = []
    for row in table.find_all("tr"):
        category_col = row.find(["td", "th"], colspan="2")
        if category_col:
            cat_name = category_col.get_text(strip=True)
            if cat_name:
                items.append(f"\n[{cat_name}]")
            continue
        
        cols = row.find_all("td")
        if cols:
            clean_name = cols[0].get_text(strip=True)
            if clean_name:
                items.append(clean_name)
    
    return items
```

```python
member_types = get_cpp_items("Member types")
member_funcs = get_cpp_items("Member functions")
non_member_funcs = get_cpp_items("Non-member functions")

print("РЕЗУЛЬТАТИ:")

print("\n Member types:")
print(", ".join(member_types) if member_types else "Не знайдено")

print("\n Member functions:")
print(", ".join(member_funcs) if member_funcs else "Не знайдено")

print("\n Non-member functions:")
print(", ".join(non_member_funcs) if non_member_funcs else "Не знайдено")
```

```python
url_ctor = 'https://en.cppreference.com/w/cpp/utility/any/any'
r_ctor = requests.get(url_ctor)
print(f"Статус сторінки конструкторів: {r_ctor.status_code}")

soup_ctor = BSoup(r_ctor.text, 'html.parser')
```

```python
ctor_table = soup_ctor.find("table", class_="t-dcl-begin")

constructors = []

if ctor_table:
    for row in ctor_table.find_all("tr", class_="t-dcl"):
        mark = row.find("span", class_="t-mark-rev")
        if not mark:
            mark = row.find("span", class_="t-mark")
        
        std_text = mark.get_text(strip=True) if mark else "(since C++98)"
        
        if mark:
            mark.extract()
        code_text = " ".join(row.get_text(strip=True).split())
        
        removed_versions = ["until C++11", "until C++14", "until C++17", "until C++20", "until C++23"]
        is_cpp23 = True
        for version in removed_versions:
            if version in std_text:
                is_cpp23 = False
                break
        
        constructors.append({
            "code": code_text,
            "standards": std_text,
            "is_cpp23": is_cpp23
        })
    
    print(f"\n Загальна кількість конструкторів: {len(constructors)}")
    
else:
    print(" Не знайдено таблицю з конструкторами")
```

```python
if constructors:
    print("ВСІ КОНСТРУКТОРИ ТА ЇХ СТАНДАРТИ:")
    
    for i, c in enumerate(constructors, 1):
        print(f"\n{i}. Стандарти: {c['standards']}")
        print(f" Код: {c['code'][:150]}...")
```

```python
if constructors:
    print("КОНСТРУКТОРИ, ЩО ІСНУЮТЬ В C++23:")
    
    cpp23_ctors = [c for c in constructors if c['is_cpp23']]
    
    if cpp23_ctors:
        for i, c in enumerate(cpp23_ctors, 1):
            print(f"\n{i}. {c['code'][:150]}...")
    else:
        print("Немає конструкторів, що існують в C++23")
```

```python
print("ПІДСУМОК:")
print(f"• Member types: {len(member_types)} елементів")
print(f"• Member functions: {len(member_funcs)} елементів")
print(f"• Non-member functions: {len(non_member_funcs)} елементів")
print(f"• Конструкторів всього: {len(constructors) if constructors else 0}")
print(f"• Конструкторів C++23: {len([c for c in constructors if c['is_cpp23']]) if constructors else 0}")
```
